```python
from textblob import TextBlob

sub = []
pol = []
for txt in df['text']:
    blob = TextBlob(txt).sentiment
    sub.append(blob.subjectivity)
    pol.append(blob.polarity)

df['subjectivity'] = sub
df['polarity'] = pol
df.head()
```
